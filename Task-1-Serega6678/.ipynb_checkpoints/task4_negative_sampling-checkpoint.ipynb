# Assignment 1.4: Negative sampling (15 points)

You may have noticed that word2vec is really slow to train. Especially with big (> 50 000) vocabularies. Negative sampling is the solution.

The task is to implement word2vec with negative sampling.

This is what was discussed in Stanford lecture. The main idea is in the formula:

$$ L = \log\sigma(u^T_o \cdot u_c) + \sum^k_{i=1} \mathbb{E}_{j \sim P(w)}[\log\sigma(-u^T_j \cdot u_c)]$$

Where $\sigma$ - sigmoid function, $u_c$ - central word vector, $u_o$ - context (outside of the window) word vector, $u_j$ - vector or word with index $j$.

The first term calculates the similarity between positive examples (word from one window)

The second term is responsible for negative samples. $k$ is a hyperparameter - the number of negatives to sample.
$\mathbb{E}_{j \sim P(w)}$
means that $j$ is distributed accordingly to unigram distribution.

Thus, it is only required to calculate the similarity between positive samples and some other negatives. Not across all the vocabulary.

Useful links:
1. [Efficient Estimation of Word Representations in Vector Space](https://arxiv.org/pdf/1301.3781.pdf)
1. [Distributed Representations of Words and Phrases and their Compositionality](http://papers.nips.cc/paper/5021-distributed-representations-of-words-and-phrases-and-their-compositionality.pdf)

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from nltk.tokenize import RegexpTokenizer
from collections import Counter
from matplotlib import pyplot as plt
import numpy as np
device = torch.device('cuda')
```

```python
tokenizer = RegexpTokenizer("[\w']+")
text = []
with open('text8', 'r') as context_file:
    text = context_file.read()
text = text.split(' ')
for cnt in range(text.count('')):
    text.remove('')
text_cnt = Counter(text)
vocabulary = set([word for (word, length) in text_cnt.most_common(200)])
vocabulary_size = len(vocabulary)
for i in range(len(text)):
    if text[i] not in vocabulary:
        text[i] = 'UNKNOWN'
ind2word = [word for word in set(text)]
word2ind = dict([[word, idx] for idx, word in enumerate(set(text))])
vocabulary_size += 1
```

```python
prob_assistant = Counter(text)
probabilities = torch.tensor([float(prob_assistant[ind2word[i]]) ** 0.75 \
                              for i in range(vocabulary_size)])
```

```python
def get_context_and_labels(text, window_size=2):
    context = []
    labels = []
    for i in range (window_size, len(text) - window_size):
        cur_neighbours = text[i - window_size : i] + text[i + 1 : i + window_size + 1]
        for elem in cur_neighbours:
            context.append(word2ind[elem])
        for j in range(window_size * 2):
            labels.append(word2ind[text[i]])
    return (context, labels)
```

```python
class CBOW_dataset(Dataset):
    def __init__(self, text):
        self.context, self.correct_words = get_context_and_labels(text)
    
    def __len__(self):
        return len(self.context)
    
    def __getitem__(self, idx):
        return [torch.tensor(self.context[idx], dtype=torch.long, device=device), \
                torch.tensor(self.correct_words[idx], dtype=torch.long, device=device)]
```

```python
train_dataset = CBOW_dataset(text[:1000000]) # to speed up computations
loader_train = DataLoader(train_dataset)
```

```python
class Neg_Sampling(nn.Module):
    def __init__(self, hidden_layer_size, vocabulary_size):
        super().__init__()
        self.emb1 = nn.Embedding(vocabulary_size, hidden_layer_size, sparse=True)
        self.emb2 = nn.Embedding(vocabulary_size, hidden_layer_size, sparse=True)
        self.vocabulary_size = vocabulary_size
        self.hidden_layer_size = hidden_layer_size
        nn.init.kaiming_normal_(self.emb1.weight)
        nn.init.kaiming_normal_(self.emb2.weight)

    def forward(self, center_idx, context_idx, neg_sample_idx):
        v = self.emb2(context_idx)
        u = self.emb1(center_idx)
        pos_score = torch.sum(torch.mul(v, u), dim=1)
        pos_score = -F.logsigmoid(pos_score)

        neg = self.emb2(neg_sample_idx)
        neg_score = torch.bmm(neg.unsqueeze(0), u.unsqueeze(2)).squeeze()
        neg_score = -torch.sum(F.logsigmoid(-neg_score), dim=0)

        total_score = torch.mean(neg_score + pos_score)
        return total_score
```

```python
trainning_info = [(0, 0)]

def train_Simple_CBOW(model, optimizer, neg_samples=5, epochs=1):
    M = torch.distributions.multinomial.Multinomial(neg_samples, probabilities)
    model = model.to(device=device)
    values = np.array([])
    for epoch in range(epochs):
        for t, (x, y) in enumerate(loader_train):
            model.train()
            samples = M.sample()
            neg_idx = []
            for idx, elem in enumerate(samples):
                while elem > 0:
                    elem -= 1
                    neg_idx.append(idx)
            neg_idx = torch.tensor(neg_idx, dtype=torch.long, device=device)
            loss = model(y, x, neg_idx)
            values = np.append(values, [loss.item()])
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            if t % 10000 == 0 and t != 0:
                trainning_info.append((t, np.mean(values)))
                values = np.array([])
            if t % 100000 == 0:
                print('Phase', t, 'out of', train_dataset.__len__())
```

```python
hidden_layer_size = 100
learning_rate = 1e-2
model = Neg_Sampling(hidden_layer_size, vocabulary_size)
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

train_Simple_CBOW(model, optimizer)
```

```python
plt.plot([a[0] for a in trainning_info], [a[1] for a in trainning_info])
```

```python
total_coord = 0.5 * (model.emb1.weight.data + model.emb2.weight.data)
```

```python
from sklearn.decomposition import PCA

my_pca = PCA(n_components=2)
```

```python
new_coord = my_pca.fit_transform(total_coord.to(device='cpu'))
```

```python
import seaborn as sns
from matplotlib.pyplot import figure

figure(figsize=(16, 12))

ax = sns.scatterplot(x=[a[0] for a in new_coord], y=[a[1] for a in new_coord])
for i in range(vocabulary_size):
    ax.text(new_coord[i][0]+.02, new_coord[i][1]+.02, ind2word[i])
```

As we can see everything works correctly - digits are near digits, synonyms are near synonyms
